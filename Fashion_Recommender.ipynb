```python
import string
import re
import nltk

# import word_tokenize & FreqDist dari NLTK
nltk.download('punkt')
from nltk.tokenize import word_tokenize

# menghapus tab, new line, backslice, non ASCII dan incomplete URL
fashion['soup'] = fashion['soup'].apply(lambda x: x.replace('\\t'," ").replace('\\n',"").replace('\\u'," ").replace('\\',"").replace("http://", " ").replace("https://"," "))
fashion['soup'] = fashion['soup'].apply(lambda x: x.encode('ascii', 'replace').decode('ascii'))
fashion['soup'] = fashion['soup'].apply(lambda x: ' '.join(re.sub("([@#][A-Za-z0-9]+)|(\w+:\/\/\S+)"," ", x).split()))

# menghapus angka
fashion['soup'] = fashion['soup'].apply(lambda x: re.sub(r"\d+", "", x))
#replace simbol with space
fashion['soup'] = fashion['soup'].apply(lambda x: x.translate({ord(c): " " for c in "!@#$%^&*()[]{};:,./<>?\|`~-=_+—"}))
# menghapus tanda baca
fashion['soup'] = fashion['soup'].apply(lambda x: x.translate(str.maketrans("","",string.punctuation)))
# menghapus whitespace leading & trailing
fashion['soup'] = fashion['soup'].apply(lambda x: x.strip())

# mengganti multiple whitespace menjadi single whitespace
fashion['soup'] = fashion['soup'].apply(lambda x: re.sub('\s+',' ', x))
# menghapus single char
fashion['soup'] = fashion['soup'].apply(lambda x: re.sub(r"\b[a-zA-Z]\b", "", x))
# NLTK word tokenize
fashion['soup'] = fashion['soup'].apply(lambda x: word_tokenize(x))
```

```python
def get_recommendations(nama):
  # Creates Series
  nama = pd.Series(nama)

  # Cleaning
  nama = nama.apply(lambda x: x.replace('\\t'," ").replace('\\n'," ").replace('\\u'," ").replace('\\',"").replace("http://", " ").replace("https://", " "))
  nama = nama.apply(lambda x: x.encode('ascii', 'replace').decode('ascii'))
  nama = nama.apply(lambda x: ' '.join(re.sub("([@#][A-Za-z0-9]+)|(\w+:\/\/\S+)"," ", x).split()))
  nama = nama.apply(lambda x: re.sub(r"\d+", "", x))
  nama = nama.apply(lambda x: x.translate({ord(c): " " for c in "!@#$%^&*()[]{};:,./<>?\|`~-=_+—"}))
  nama = nama.apply(lambda x: x.strip())
  nama = nama.apply(lambda x: re.sub('\s+',' ', x))
  nama = nama.apply(lambda x: re.sub(r"\b[a-zA-Z]\b", "", x))

  # Tokenizing
  nama = nama.apply(lambda x: word_tokenize(x))

  # Filtering
  from nltk.corpus import stopwords
  # memeriksa daftar kata di stopwords
  eng = stopwords.words('english')
  # list stopwords bahasa inggris
  list_stopwords = eng
  # hapus stopword pada list token
  def stopwords_removal(words):
    return [word for word in words if word not in list_stopwords]

  nama = nama.apply(stopwords_removal)

  # mengubah list pada kolom soup menjadi string
  nama = nama.agg(lambda x:' '.join(map(str, x)))

  # Ambil kata kunci
  nama = nama.to_string()
  nama = nama[6:-1]
  nama = re.sub(r'[^\w\s]', '', nama)

  # Membaca data
  fashion = pd.read_csv("data_soup.csv")
  brnd = "['keywords', {nama}]"
  data = [{'Unnamed: 0': 2192,'title': nama, 'brand': brnd, 'description': nama,
           'category': nama, 'price': nama, 'color': nama, 'image_url': nama, 'url': nama,
           'title_list': nama, 'brand_list': nama, 'description_list': nama,
           'category_list': nama, 'color_list': nama, 'soup': nama}]

  fashion.drop_duplicates()

  # Menggabungkan database buku dengan kata kunci yang dimasukkan
  new = pd.DataFrame(data)
  df = [fashion, new]
  result = pd.concat(df)
  fashion = result.copy()

  # TF-IDF Vectorizer
  tfidf = TfidfVectorizer(stop_words=list_stopwords)
  tfidf_matrix = tfidf.fit_transform(fashion['soup'])

  # Menghitung matrix cosine similarity
  cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

  fashion = fashion.reset_index()
  titles = fashion['title']
  titles = titles.str.lower()
  color = fashion['color']
  category = fashion['category']
  description = fashion['description']

  # mengubah string ke list
  fashion['brand_list'] = fashion['brand'].apply(lambda x:x.replace("'", "").replace("[","").replace("]", ""))
  fashion['brand_list'] = fashion.brand_list.apply(lambda x: x.split(', '))

  # list brand ke dua
  brn = []
  for brand in fashion['brand_list']:
    brn.append(brand[0])

  # list title
  tit = []
  for title in fashion['title']:
    tit.append(title)

  # hapus brand kedua jika sama dengan title
  x = 0
  for i in fashion['brand_list']:
    if i[0] == tit[x]:
      fashion['brand_list'][x] = fashion['brand_list'][x][0]
      x = x + 1

  # set produk yang memiliki 1 brand menjadi list
  y = 0
  for i in fashion['brand_list']:
    if isinstance(i, str):
      brand['brand_list'][y] = i.split("delimiter")
      y = y + 1

  fashion['brand_list'] = [', '.join(map(str, l)) for l in fashion['brand_list']]

  brands = fashion['brand_list']
  thumbnail = fashion['url'].apply(lambda x: '<img src="{}" width="226px"/>'.format(x) if x else '')
  indices = pd.Series(fashion.index, index=titles)

  # Memebentuk tampilan dataframe berdasarkan judul, penulis, genre dan cover
  new_rec = pd.DataFrame(columns = ['Title', 'Brand', 'Category', 'Description', 'Color', 'Thumbnail'])
  count = 0

  idx = indices[nama]
  sim_scores = list(enumerate(cosine_sim[idx]))
  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
  sim_scores = sim_scores[1:6]
  fashion_indices = [i[0] for i in sim_scores]

  for i in fashion_indices:
    new_rec.at[count+1, 'Title'] = titles.iloc[fashion_indices[count]].title()
    new_rec.at[count+1, 'Brand'] = brands.iloc[fashion_indices[count]].title()
    new_rec.at[count+1, 'Category'] = category.iloc[fashion_indices[count]].title()
    new_rec.at[count+1, 'Description'] = description.iloc[fashion_indices[count]].title()
    new_rec.at[count+1, 'Color'] = color.iloc[fashion_indices[count]].title()
    new_rec.at[count+1, 'Thumbnail'] = thumbnail.iloc[fashion_indices[count]]
    count += 1

  return new_rec
```
